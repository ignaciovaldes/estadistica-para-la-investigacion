## Ejercicio para entregar 1
Calcula la correlacin de las variables X,Y en el ejercicio de las plizas de seguro.

---

```python
import numpy as np
import pandas as pd

# Valores de X, P(X)
x_values = np.array([100, 500, 1000])
px_values = np.array([0.35, 0.40, 0.25])

# Valores de Y, P(Y)
y_values = np.array([500, 1000, 5000])
py_values = np.array([0.55, 0.35, 0.10])

# Tabla de probabilidad conjunta P(X,Y)
p_xy = np.array([
    [0.30, 0.05, 0.00],
    [0.15, 0.20, 0.05],
    [0.10, 0.10, 0.05]
])

# Calcular E(X)
E_X = np.sum(x_values * px_values)

# Calcular E(Y)
E_Y = np.sum(y_values * py_values)

# Calcular E(X^2)
E_X2 = np.sum((x_values ** 2) * px_values)

# Calcular E(Y^2)
E_Y2 = np.sum((y_values ** 2) * py_values)

# Calcular Var(X) y Var(Y)
Var_X = E_X2 - E_X**2
Var_Y = E_Y2 - E_Y**2

# Calcular sigma_X y sigma_Y
sigma_X = np.sqrt(Var_X)
sigma_Y = np.sqrt(Var_Y)

# Calcular E(XY)
E_XY = 0
for i in range(len(x_values)):
    for j in range(len(y_values)):
        E_XY += x_values[i] * y_values[j] * p_xy[i, j]

# Calcular Cov(X, Y)
cov_XY = E_XY - E_X * E_Y

# Calcular la correlacin rho_XY
rho_XY = cov_XY / (sigma_X * sigma_Y)

print("E(X) =", E_X)
print("E(Y) =", E_Y)
print("Var(X) =", Var_X)
print("Var(Y) =", Var_Y)
print("sigma_X =", sigma_X)
print("sigma_Y =", sigma_Y)
print("Cov(X, Y) =", cov_XY)
print("Correlacin (rho_XY) =", rho_XY)
```

```python
import sympy as sp

# Definir las variables
x, y = sp.symbols('x y')

# Definir la fdp conjunta f(x, y)
f_xy = 24 * x * y

# Definir los lmites del dominio
x_min, x_max = 0, 1
y_min = 0
y_max = 1 - x

# Calcular E(X)
E_X = sp.integrate(x * f_xy, (y, y_min, y_max), (x, x_min, x_max))

# Calcular E(Y)
E_Y = sp.integrate(y * f_xy, (y, y_min, y_max), (x, x_min, x_max))

# Calcular E(X^2)
E_X2 = sp.integrate(x**2 * f_xy, (y, y_min, y_max), (x, x_min, x_max))

# Calcular E(Y^2)
E_Y2 = sp.integrate(y**2 * f_xy, (y, y_min, y_max), (x, x_min, x_max))

# Calcular Var(X) y Var(Y)
Var_X = E_X2 - E_X**2
Var_Y = E_Y2 - E_Y**2

# Calcular sigma_X y sigma_Y
sigma_X = sp.sqrt(Var_X)
sigma_Y = sp.sqrt(Var_Y)

# Calcular E(XY)
E_XY = sp.integrate(x * y * f_xy, (y, y_min, y_max), (x, x_min, x_max))

# Calcular Cov(X, Y)
Cov_XY = E_XY - E_X * E_Y

# Calcular la correlacin rho_XY
rho_XY = Cov_XY / (sigma_X * sigma_Y)

# Mostrar los resultados
print("E(X) =", E_X)
print("E(Y) =", E_Y)
print("Var(X) =", Var_X)
print("Var(Y) =", Var_Y)
print("sigma_X =", sigma_X)
print("sigma_Y =", sigma_Y)
print("Cov(X, Y) =", Cov_XY)
print("Correlacin (rho_XY) =", rho_XY)
```

## Ejercicio para entregar 2

```python
# Importar las libreras necesarias
import numpy as np
from scipy import stats

# Datos de vida til en horas para cada modelo
modelo1 = [8090, 8090, 8120, 8111, 8112, 8115, 8126, 8116, 8108, 8112]
modelo2 = [7994, 8115, 7810, 8110, 7999, 7811, 8050, 8068, 7956, 8106]

# Calcular las medias de cada modelo
media1 = np.mean(modelo1)
media2 = np.mean(modelo2)

print(f"Media Modelo 1: {media1:.2f} horas")
print(f"Media Modelo 2: {media2:.2f} horas")

# Calcular las varianzas y desviaciones estndar de cada modelo
varianza1 = np.var(modelo1, ddof=1)
varianza2 = np.var(modelo2, ddof=1)

desv_est1 = np.sqrt(varianza1)
desv_est2 = np.sqrt(varianza2)

print(f"Desviacin estndar Modelo 1: {desv_est1:.2f} horas")
print(f"Desviacin estndar Modelo 2: {desv_est2:.2f} horas")

# Realizar el ANOVA de una va
F_stat, p_value = stats.f_oneway(modelo1, modelo2)

print(f"Estadstico F: {F_stat:.2f}")
print(f"Valor p: {p_value:.4f}")

# Grados de libertad
n1 = len(modelo1)
n2 = len(modelo2)
df_between = 1
df_within = n1 + n2 - 2

# Suma de cuadrados dentro de los grupos (SSW)
SSW = (n1 - 1) * varianza1 + (n2 - 1) * varianza2
MSW = SSW / df_within

print(f"Suma de cuadrados dentro de los grupos (SSW): {SSW:.2f}")
print(f"Cuadrados medios dentro de los grupos (MSW): {MSW:.2f}")

# Valor crtico de t al 5% de significancia
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df_within)

print(f"Valor crtico de t: {t_crit:.3f}")

# Clcular LSD
LSD = t_crit * np.sqrt(2 * MSW / n1)

print(f"LSD al 5%: {LSD:.2f} horas")

# Diferencia de medias
diferencia_medias = abs(media1 - media2)

print(f"Diferencia de medias: {diferencia_medias:.2f} horas")

# Comparacin de la diferencia de medias con el LSD
if diferencia_medias > LSD:
    print("La diferencia es significativa al nivel del 5%.")
    print("Se recomienda seleccionar el Modelo 1 por su mayor vida til promedio.")
else:
    print("No hay diferencia significativa al nivel del 5%.")
```

## Ejercicio para entregar 3

```python
# Importar las libreras necesarias
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from itertools import combinations

# Crear el DataFrame con los datos proporcionados
data = {
    'Cantidad': ['X']*12 + ['Y']*12,
    'Formato': ['Barra']*4 + ['Granulado']*4 + ['Polvo']*4 + ['Barra']*4 + ['Granulado']*4 + ['Polvo']*4,
    'Piezas': [
        28, 26, 26, 24,  # Cantidad X - Barra
        31, 29, 27, 27,  # Cantidad X - Granulado
        30, 31, 29, 28,  # Cantidad X - Polvo
        20, 22, 24, 25,  # Cantidad Y - Barra
        25, 23, 25, 26,  # Cantidad Y - Granulado
        27, 28, 27, 29   # Cantidad Y - Polvo
    ]
}

df = pd.DataFrame(data)

# Mostrar el DataFrame
print("Datos:")
print(df)

# Realizar el ANOVA de dos vas con interaccin
model = ols('Piezas ~ C(Cantidad) * C(Formato)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("\nTabla ANOVA:")
print(anova_table)

# Extraer el Mean Square Error (MSE)
MSE = anova_table['sum_sq']['Residual'] / anova_table['df']['Residual']
print(f"\nMean Square Error (MSE): {MSE:.4f}")

# Valor crtico de t al 95% de confianza
alpha = 0.05
df_error = anova_table['df']['Residual']
t_crit = stats.t.ppf(1 - alpha/2, df_error)
print(f"Valor crtico de t (gl={df_error}): {t_crit:.4f}")

# Clcular del LSD para el Factor A (Cantidad)
n_A = df['Cantidad'].value_counts()[0]  # Nmero de observaciones por nivel de Cantidad
SE_A = np.sqrt(2 * MSE / n_A)
LSD_A = t_crit * SE_A
print(f"\nLSD para el Factor A (Cantidad): {LSD_A:.4f}")

# Clcular del LSD para el Factor B (Formato)
n_B = df['Formato'].value_counts()[0]  # Nmero de observaciones por nivel de Formato
SE_B = np.sqrt(2 * MSE / n_B)
LSD_B = t_crit * SE_B
print(f"LSD para el Factor B (Formato): {LSD_B:.4f}")

# Clcular del LSD para la interaccin (Cantidad x Formato)
n_interaction = df.groupby(['Cantidad', 'Formato']).size().iloc[0]  # Nmero de observaciones por combinacin
SE_interaction = np.sqrt(2 * MSE / n_interaction)
LSD_interaction = t_crit * SE_interaction
print(f"LSD para la interaccin (Cantidad x Formato): {LSD_interaction:.4f}")

# Medias de cada nivel de Factor A y B
mean_Cantidad = df.groupby('Cantidad')['Piezas'].mean()
mean_Formato = df.groupby('Formato')['Piezas'].mean()

print("\nMedias por Cantidad:")
print(mean_Cantidad)
print("\nMedias por Formato:")
print(mean_Formato)

# Medias de la interaccin (Cantidad x Formato)
mean_interaction = df.groupby(['Cantidad', 'Formato'])['Piezas'].mean()
mean_interaction = mean_interaction.reset_index()
print("\nMedias de la interaccin (Cantidad x Formato):")
print(mean_interaction)

# Comparar las medias para el Factor A (Cantidad)
diff_A = abs(mean_Cantidad['X'] - mean_Cantidad['Y'])
print(f"\nDiferencia de medias (Cantidad X - Cantidad Y): {diff_A:.4f}")

if diff_A > LSD_A:
    print(f"La diferencia es significativa (p < 0.05): {diff_A:.4f} > {LSD_A:.4f}")
else:
    print(f"No hay diferencia significativa: {diff_A:.4f} <= {LSD_A:.4f}")

# Comparaciones pareadas para el Factor B (Formato)
print("\nComparaciones pareadas para el Factor B (Formato):")
formatos = mean_Formato.index.tolist()
for f1, f2 in combinations(formatos, 2):
    diff_B = abs(mean_Formato[f1] - mean_Formato[f2])
    print(f"Diferencia de medias ({f1} - {f2}): {diff_B:.4f}")
    if diff_B > LSD_B:
        print(f"  La diferencia es significativa (p < 0.05): {diff_B:.4f} > {LSD_B:.4f}")
    else:
        print(f"  No hay diferencia significativa: {diff_B:.4f} <= {LSD_B:.4f}")

# Comparaciones pareadas para la interaccin (Cantidad x Formato)
print("\nComparaciones pareadas para la interaccin (Cantidad x Formato):")
interactions = mean_interaction[['Cantidad', 'Formato', 'Piezas']].values.tolist()
for i1, i2 in combinations(interactions, 2):
    group1 = f"{i1[0]} - {i1[1]}"
    group2 = f"{i2[0]} - {i2[1]}"
    diff_interaction = abs(i1[2] - i2[2])
    print(f"Diferencia de medias ({group1} vs {group2}): {diff_interaction:.4f}")
    if diff_interaction > LSD_interaction:
        print(f"  La diferencia es significativa (p < 0.05): {diff_interaction:.4f} > {LSD_interaction:.4f}")
    else:
        print(f"  No hay diferencia significativa: {diff_interaction:.4f} <= {LSD_interaction:.4f}")
```
